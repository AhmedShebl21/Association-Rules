```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_excel("q2.xlsx")
```

```python
data.head(10)
```

```python
q1 = pd.read_excel('q2.xlsx', header=None)
```

```python
q1
```

```python
q1.drop(0,axis=1,inplace=True)
```

```python
q1
```

```python
records = []
for i in range(0, 9):
    records.append([str(q1.values[i,j]) for j in range(0, 4)])
```

```python
records
```

```python
te = TransactionEncoder()
te_ary=te.fit(records).transform(records)
te_ary
```

```python
df=pd.DataFrame(te_ary,columns=te.columns_)
```

```python
df
```

```python
df.drop(['nan'], axis=1 , inplace =True)
```

```python
df
```

```python
freq_itemsets=apriori(df,min_support=.22,use_colnames=True)
freq_itemsets
```

```python
rules = association_rules(freq_itemsets, metric ="confidence",min_threshold=.6)
rules
```

```python
assoc_rules = rules[['antecedents','consequents','antecedent support','consequent support','support','confidence','lift']]
assoc_rules
```
